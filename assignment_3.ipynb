**Question 1**
Plot using sns.relplot, line plot, box plot on fmri data set

```python
import numpy as np
import pandas as pd
import seaborn as sns

import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = sns.load_dataset('fmri')
data.head()
```

```python
data.info()
```

```python
sns.relplot(x = 'timepoint', y = 'signal', data = data, color = 'orange')
```

```python
sns.relplot(x = 'timepoint', y = 'signal', hue ='region', kind = 'line', data = data)
```

```python
sns.relplot(x = 'timepoint', y = 'signal', kind = 'line',style = 'event', data = data)
```

```python
sns.relplot(x='timepoint',y='signal',data=data, hue='region',kind='line',style='event')
```

```python
sns.relplot(x='timepoint',y='signal',data=data, kind = 'scatter',hue='event',style='region')
```
